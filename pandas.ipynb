# **4. Cleaning**

## 4.1. Column names

```python
d.rename(columns = {"Reporting year":"year"}, inplace = True)
d.rename(columns = {"Region":"region"}, inplace = True)
d.rename(columns = {"Survey comparability":"survey comparability"}, inplace = True)
d.rename(columns = {"Monetary (%)":"monetary"}, inplace = True)
d.rename(columns = {"Educational attainment (%)":"ea"}, inplace = True)
d.rename(columns = {"Educational enrollment (%)":"ee"}, inplace = True)
d.rename(columns = {"Electricity (%)":"electricity"}, inplace = True)
d.rename(columns = {"Sanitation (%)":"sanitation"}, inplace = True)
d.rename(columns = {"Drinking water (%)":"dwater"}, inplace = True)
d.rename(columns = {"Multidimensional poverty headcount ratio (%)":"multidimensional ratio"}, inplace = True)
```

```python
d2=d
```

## 4.2. Delete columns

```python
d2 = d2.drop(columns = ["Survey coverage","Welfare type","Survey name","survey comparability"])
```

```python
d2.head()
```

## 4.3. Drop NA

```python
d2 = d2.replace("-",np.NaN)
d2.head()
```

```python
d2 =d2.dropna()
```

```python
d2.head()
```

# **5. Indexing**

```python
d2 = d2.set_index(["Economy","year"])
d2.head()
```

```python
import pandas as pd

# Define una función personalizada para el formato
def custom_format(x):
    if isinstance(x, float):
        return '{:.3f}'.format(x)
    return str(x)

# Aplica la función de formato personalizada a todos los datos del DataFrame
pd.options.display.float_format = custom_format

# Ejecuta tu código
d2[["ea", "ee", "electricity", "sanitation", "dwater", "multidimensional ratio"]].describe()
```

```python
d2[["ea", "ee" , "sanitation", "dwater", "multidimensional ratio"]].describe().applymap(lambda x:"{:,.1f}".format(x))
```
